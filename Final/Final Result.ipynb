```python
data = {
    'NN_SMOTE' : NN_SMOTE,
    'NN' : NN,
    'RF_SMOTE' : RF_SMOTE,
    'DT_SMOTE' : DT_SMOTE,
    'NB_SMOTE' : NB_SMOTE,
    'LR_SMOTE' : LR_SMOTE,
    'KNN_SMOTE' : KNN_SMOTE
}
data
```

```python
data['NN_SMOTE']['KC1']['Train']['Gmeans']
```

```python


dataset = []
alg = []

train_gmeans = []
train_recall = []
train_pf = []

dev_gmeans = []
dev_recall = []
dev_pf = []

test_gmeans = []
test_recall = []
test_pf = []


for i in data: #Alg
    for j in data[i]: #dataset
        for k in data[i][j]: #set
            for l in data[i][j][k]: #metric
                if(l != 'cf_matrix'):
                    if(k == 'Train' and l == 'Gmeans'):
                        alg.append(i)
                        dataset.append(j)
                        
                        train_gmeans.append(data[i][j][k][l])
                    if(k == 'Train' and l == 'Recall'):
                        train_recall.append(data[i][j][k][l])
                    if(k == 'Train' and l == 'pf'):
                        train_pf.append(data[i][j][k][l])
                        
                    if(k == 'Dev' and l == 'Gmeans'):       
                        dev_gmeans.append(data[i][j][k][l])
                    if(k == 'Dev' and l == 'Recall'):
                        dev_recall.append(data[i][j][k][l])
                    if(k == 'Dev' and l == 'pf'):
                        dev_pf.append(data[i][j][k][l])
                        
                    if(k == 'Test' and l == 'Gmeans'):
                        test_gmeans.append(data[i][j][k][l])
                    if(k == 'Test' and l == 'Recall'):
                        test_recall.append(data[i][j][k][l])
                    if(k == 'Test' and l == 'pf'):
                        test_pf.append(data[i][j][k][l])
```

```python
dataset_ = ['KC1', 'KC2', 'PC1', 'PC3', 'PC4', 'CM1', 'JM1']
alg_ = ['NN_SMOTE', 'NN', 'rf', 'dt', 'nb', 'lr', 'knn']
hyperparam = []
for i in alg_:
    for j in dataset_:
        if(i == 'NN_SMOTE'):
            hyperparam.append(globals()['best_hyperparam_' + j])
        elif(i == 'NN'):
            hyperparam.append(globals()['best_hyperparam_'+ j +'_without_SMOTE'])
        else:
            hyperparam.append(globals()['best_hyperparam_'+ i +'_' + j])
```

```python
# all_data['dataset'] = ['NN_SMOTE', 'NN', 'RF_SMOTE', 'DT_SMOTE', 'NB_SMOTE', 'LR_SMOTE', 'KNN_SMOTE']
all_data = {}
all_data['dataset'] = dataset
all_data['alg'] = alg
all_data['hyperparam'] = hyperparam
all_data['train_gmeans'] = train_gmeans
all_data['train_recall'] = train_recall
all_data['train_pf'] = train_pf

all_data['dev_gmeans'] = dev_gmeans
all_data['dev_recall'] = dev_recall
all_data['dev_pf'] = dev_pf


all_data['test_gmeans'] = test_gmeans
all_data['test_recall'] = test_recall
all_data['test_pf'] = test_pf
```

```python
df = pd.DataFrame(all_data)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df)
```

```python
for i in df2.index:
#     NN
    if ('drop' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['drop'] = round(df2.loc[i]['hyperparam']['drop'],2)
        df2.loc[i]['hyperparam']['Dropout'] = df2.loc[i]['hyperparam'].pop('drop')
        
    if ('layer' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['layer'] = int(df2.loc[i]['hyperparam']['layer'])
        df2.loc[i]['hyperparam']['Hidden Layer'] = df2.loc[i]['hyperparam'].pop('layer')
        
    if ('lr' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['lr'] = round(df2.loc[i]['hyperparam']['lr'],5)
        df2.loc[i]['hyperparam']['Learning Rate'] = df2.loc[i]['hyperparam'].pop('lr')
        
    if ('neurons' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['neurons'] = int(df2.loc[i]['hyperparam']['neurons'])
        df2.loc[i]['hyperparam']['Neurons'] = df2.loc[i]['hyperparam'].pop('neurons')

#         SMOTE
        
    if ('neighbors' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['neighbors'] = int(df2.loc[i]['hyperparam']['neighbors'])
        df2.loc[i]['hyperparam']['Neighbors'] = df2.loc[i]['hyperparam'].pop('neighbors')
        
    if ('sampling' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['sampling'] = round(df2.loc[i]['hyperparam']['sampling'], 2)
        df2.loc[i]['hyperparam']['Sampling'] = df2.loc[i]['hyperparam'].pop('sampling')
        
#         Random Forest
    if ('depth' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['depth'] = int(df2.loc[i]['hyperparam']['depth'])
        df2.loc[i]['hyperparam']['Max Depth'] = df2.loc[i]['hyperparam'].pop('depth')
        
    if ('estimator' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['estimator'] = int(df2.loc[i]['hyperparam']['estimator'])
        df2.loc[i]['hyperparam']['N Estimator'] = df2.loc[i]['hyperparam'].pop('estimator')

    if ('max_leaf_nodes' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['max_leaf_nodes'] = int(df2.loc[i]['hyperparam']['max_leaf_nodes'])
        df2.loc[i]['hyperparam']['Max Leaf Nodes'] = df2.loc[i]['hyperparam'].pop('max_leaf_nodes')
        
    if ('max_features' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['max_features'] = round(df2.loc[i]['hyperparam']['max_features'],2)
        df2.loc[i]['hyperparam']['Max Features'] = df2.loc[i]['hyperparam'].pop('max_features')
        
    if ('min_leaf' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['min_leaf'] = int(df2.loc[i]['hyperparam']['min_leaf'])
        df2.loc[i]['hyperparam']['Min Samples Leaf'] = df2.loc[i]['hyperparam'].pop('min_leaf')
        
    if ('min_split' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['min_split'] = int(df2.loc[i]['hyperparam']['min_split'])
        df2.loc[i]['hyperparam']['Min Samples Split'] = df2.loc[i]['hyperparam'].pop('min_split')
        
#         NB
    if ('var_smoothing' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['var_smoothing'] = round(df2.loc[i]['hyperparam']['var_smoothing'], 2)
        df2.loc[i]['hyperparam']['Var Smoothing'] = df2.loc[i]['hyperparam'].pop('var_smoothing')
        
    if ('c' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['c'] = round(df2.loc[i]['hyperparam']['c'],2)
        df2.loc[i]['hyperparam']['C'] = df2.loc[i]['hyperparam'].pop('c')
        
    if ('max_iter' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['max_iter'] = int(df2.loc[i]['hyperparam']['max_iter'])
        df2.loc[i]['hyperparam']['Max Iter'] = df2.loc[i]['hyperparam'].pop('max_iter')

# KNN
    if ('n' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['n'] = int(df2.loc[i]['hyperparam']['n'])
        df2.loc[i]['hyperparam']['n Neighbors'] = df2.loc[i]['hyperparam'].pop('n')
        
    if ('leaf' in df2.loc[i]['hyperparam']):
        df2.loc[i]['hyperparam']['leaf'] = int(df2.loc[i]['hyperparam']['leaf'])
        df2.loc[i]['hyperparam']['Leaf Size'] = df2.loc[i]['hyperparam'].pop('leaf')
```
