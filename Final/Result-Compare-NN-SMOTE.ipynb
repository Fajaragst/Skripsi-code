```python
def wilcox(data1,data2):
    data1 = data1
    data2 = data2

    stat, p = wilcoxon(data1, data2,alternative='greater')

    # interpret
    alpha = 0.05
    if p > alpha:
        print('Same distribution (fail to reject H0)')
    else:
        print('Different distribution (reject H0)')
    return(stat, p)
```
